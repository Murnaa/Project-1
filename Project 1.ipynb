#  Plotting  Data

### **Hidden Inferences are insights that are not readily apparent from the summary statistics but can be revealed through further analysis. We can use histograms,sub plots,scatter plots and heatmaps to visualize the data distribution and identify patterns.**

## **Histogram**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
data.hist(figsize=(10,10), color='blue')
plt.show()
```

## **Sub Plot**

```python
fig, axs = plt.subplots(ncols=5,nrows=3, figsize=(15, 10))
sns.kdeplot(data, x='fixed acidity', ax=axs[0,0])
sns.kdeplot(data, x='volatile acidity', ax=axs[0, 1])
sns.kdeplot(data, x='citric acid', ax=axs[0,2])
sns.kdeplot(data, x='residual sugar', ax=axs[0,3])
sns.kdeplot(data, x='chlorides', ax=axs[0,4])
sns.kdeplot(data, x='free sulfur dioxide', ax=axs[1,0])
sns.kdeplot(data, x='total sulfur dioxide', ax=axs[1, 1])
sns.kdeplot(data, x='density', ax=axs[1,2])
sns.kdeplot(data, x='pH', ax=axs[1,3])
sns.kdeplot(data, x='sulphates', ax=axs[1,4])
sns.kdeplot(data, x='alcohol', ax=axs[2, 0])
sns.kdeplot(data, x='quality', ax=axs[2, 1])
```

## **Scatter Plot**

### **They can be used to see correlations between two columns.**

```python
plt.scatter(data['quality'], data['pH'], color='green')
plt.xlabel('quality')
plt.ylabel('pH')
plt.show()
```

## **Heatmap**

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Columns')
plt.show()
```
